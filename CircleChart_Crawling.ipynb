```python
base_url = "https://circlechart.kr/page_chart/album.circle?nationGbn=T&targetTime={}&hitYear={}&termGbn=month&yearTime=3"

# Define the starting year and month
start_year = 2023
start_month = 1

# Define the number of rows you want to retrieve for each page
num_rows = 100

# Define a function to generate the next URL
def next_url(year, month):
    return base_url.format(month, year)

# Define a function to retrieve the data for a given URL
def get_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    # Extract the data from the HTML using BeautifulSoup
    data = soup.find_all("td")
    return data

# Start the loop
year = start_year
month = start_month
for i in range(num_rows):
    # Generate the URL
    url = next_url(year, month)
    # Get the data
    data = get_data(url)
    # Do something with the data
    print(year, month, data)
    # Update the year and month
    if month == 1:
        year -= 1
        month = 12
    else:
        month -= 1
```

```python
# Define the target URLs
#urls = ['https://circlechart.kr/page_chart/album.circle?nationGbn=T&targetTime=12&hitYear=2022&termGbn=month&yearTime=3']

# Define an empty list to store the data
data = []


# Start a webdriver instance
driver = wd.Chrome(executable_path='C:\\Users\\sangm\\Downloads\\chromedriver.exe')
#browser = webdriver.Chrome(options=options)

# Iterate over the URLs
for url in urls:
    # Navigate to the URL
    driver.get(url)

    # Find all the rows in the table using a CSS selector
    rows = driver.find_elements_by_css_selector('#pc_chart_tbody > tr')

    # Iterate over the rows and extract the relevant information
    for row in rows:
        cells = row.find_elements_by_css_selector('td')
        if len(cells) >= 1:
            rank = cells[0].text.strip()
            title = row.find_element_by_css_selector('div.font-bold.mb-2').text.strip()
            artist = row.find_element_by_css_selector('div.text-sm.text-gray-400.font-bold').text.strip()
            sales = driver.execute_script("return arguments[0].textContent", row.find_element_by_css_selector('#pc_chart_tbody > tr > td.text-center > span'))
            date = driver.find_element_by_xpath('/html/body/div[1]/main[2]/div/div/aside[4]/section/div[2]/label/select/option[2]').text
            
            #date = date.

    
#body > div.flex.justify-center > main.hidden.md\:block > div > div > aside.mt-14.text-\[12px\] > section > div.float-right > label > select
            # Store the information in a dictionary
            entry = {'Rank': rank, 'Title': title, 'Artist': artist, 'Sales': sales, 'Date':date}

            # Append the dictionary to the data list
            data.append(entry)

# Create a pandas DataFrame from the data list
df = pd.DataFrame(data, columns=['Rank', 'Title', 'Artist', 'Sales', 'Date'])
# Show the first 5 rows of the DataFrame
df.head()

# Quit the webdriver instance
driver.quit()
```
